```python
# Creating WQi coloum based on defined formula
```

```python
df_num_final = df1.select_dtypes(exclude="object")

df_num_final.drop(["Total Dissolved Solids", "Arsenic"], axis=1, inplace=True)

# Weight Vector(wi)
wi = np.array([0.2604,  0.4426, 0.0492, 0.0221, 0.0022, 0.1404])

# Standard values of parameters(si)
si = np.array([ 8.5, 5, 45, 100, 1000, 1.5])

# Ideal values of paramters(vIdeal)
vIdeal = np.array([ 7, 0, 0, 0, 0, 0.5])

def calc_wqi(sample):
    wqi_sample = 0
    num_col = 6
    for index in range(num_col):
        v_index = sample[index] # Obeserved value of sample at index
        v_index_ideal = vIdeal[index] # Ideal value of obeserved value
        w_index = wi[index] # weight of corresponding parameter of obeserved value
        std_index = si[index] # Standard value recommended for obeserved value
        q_index = (v_index - v_index_ideal) / (std_index - v_index_ideal)
        q_index = q_index * 100 # Final qi value of obeserved value
        wqi_sample += q_index*w_index
    return wqi_sample
    
```

```python
def calc_wqi_for_df(df1):
    wqi_arr = []
    for index in range(df1.shape[0]):
        index_row = df1.iloc[index, :]
        wqi_row = calc_wqi(index_row)
        wqi_arr.append(wqi_row)
    return wqi_arr
```

```python
wqi_arr = calc_wqi_for_df(df_num_final)
```

```python
wqi_arr_df = pd.DataFrame(wqi_arr, columns=["WQI"]).reset_index()
df1 = df1.reset_index()


```

```python

# Combining dataframe of WQI and dataframe of attributes 
df_wqi = pd.concat([df1, pd.DataFrame(wqi_arr, columns=["WQI"])], axis=1)
df_wqi.drop("index", axis=1, inplace=True)
df_wqi.shape
```

```python
#Hypothesis Testing at 50:50 of decision tree and Xg boost
```

```python
from scipy import stats

# Assuming acc1 and acc2 are the accuracy scores of two classification models
acc_model1 = 0.9885


acc_model2 = 0.9674

# Compute the difference in accuracy
diff = acc1 - acc2

# Assuming n1 and n2 are the number of samples used for each model
n1 = 731
n2 = 731

# Compute the standard error of the difference
se = ((acc1 * (1 - acc1)) / n1 + (acc2 * (1 - acc2)) / n2) ** 0.5

# Compute the z-score
z_score = diff / se

# Compute the p-value
p_value = stats.norm.sf(abs(z_score)) * 2  # two-tailed test

# Set the significance level
alpha = 0.05

# Perform hypothesis testing
if p_value < alpha:
    print("Reject the null hypothesis: The difference in accuracy is statistically significant.")
else:
    print("Fail to reject the null hypothesis: The difference in accuracy is not statistically significant.")
```

```python


# Accuracy scores for two classification models

# Significance level (alpha)

# Plot the distribution of accuracy scores
plt.figure(figsize=(8, 6))
plt.hist([acc_model1, acc_model2], bins=10,  label=['Decision tree', 'Model 2'])
plt.axvline(x=acc_model1, color='blue', linestyle='--', label='Accuracy of Decision Tree')
plt.axvline(x=acc_model2, color='green', linestyle='--', label='Accuracy of XG Boost')

# Plot the significance level
plt.axvspan(acc_model2 - alpha, acc_model2 + alpha,  alpha=0.3,color='red' ,label='Significance Level')

# Add labels and title
plt.xlabel('Accuracy')
plt.ylabel('Frequency')
plt.title('Hypothesis Testing for Classification Models')
plt.legend()

# Show plot
plt.grid(True)
plt.show()
```
